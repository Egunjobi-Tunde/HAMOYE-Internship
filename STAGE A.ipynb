```python
#Pandas can connect to databases, get data with queries and save in a dataframe.
import pandas as pd
url= 'https://github.com/WalePhenomenon/climate_change/blob/master/fuel_ferc1.csv?raw=True'
fuel_data = pd.read_csv(url, error_bad_lines= False )
fuel_data.describe(include= 'all' )
```

```python
fuel_data
```

```python
fuel_data.isnull().sum()
```

```python
fuel_data.groupby('fuel_unit')['fuel_unit'].count()
```

```python
fuel_data[[ 'fuel_unit' ]] = fuel_data[[ 'fuel_unit' ]].fillna(value= 'mcf' )
```

```python
fuel_data.isna().sum()
```

```python
fuel_data.groupby('report_year')['report_year'].count()
```

```python
len(set(fuel_data.report_year))
```

```python
f1 = fuel_data.iloc[0:19000].reset_index(drop = True)
f2 = fuel_data.iloc[19000:].reset_index(drop=True)
```

```python
assert len(fuel_data) == (len(f1) + len(f2))
```

```python
f3 = pd.merge(f1,f2, how = 'inner') #interscetion
f3
```

```python
f4 = pd.merge(f1,f2, how = 'outer') #union
f4
```

```python
f5 = pd.merge(f1,f2, how = 'left') # f1 n f2'
f5
```

```python
f6 = pd.merge(f1,f2, how = 'right') # f1' n f2
f6
```

```python
f7 = pd.concat([f1, f2]).reset_index(drop = True)
f7
```

```python
#check for duplicate rows
f7.duplicated().any()
```

```python
# Import plotting library
import matplotlib.pyplot as plt
import seaborn as sns
plt.figure(figsize=( 7 , 4 ))
plt.xticks(rotation= 90 )
fuel_unit = pd.DataFrame({ 'unit' :[ 'BBL' , 'GAL' , 'GRAMSU' , 'KGU' , 'MCF' , 'MMBTU' ,
'MWDTH' , 'MWHTH' , 'TON' ],
'count' :[ 7998 , 84 , 464 , 110 , 11354 , 180 , 95 , 100 , 8958 ]})
sns.barplot(data=fuel_unit, x= 'unit' , y= 'count' )
plt.xlabel( 'Fuel Unit' )
```

```python
# Import plotting library
#using log value for y
import matplotlib.pyplot as plt
import seaborn as sns
plt.figure(figsize=( 7 , 4 ))
plt.xticks(rotation= 90 )
fuel_unit = pd.DataFrame({ 'unit' :[ 'BBL' , 'GAL' , 'GRAMSU' , 'KGU' , 'MCF' , 'MMBTU' ,
'MWDTH' , 'MWHTH' , 'TON' ],
'count' :[ 7998 , 84 , 464 , 110 , 11354 , 180 , 95 , 100 , 8958 ]})
g = sns.barplot(data=fuel_unit, x='unit', y='count')
g.set_yscale('log')
g.set_ylim( 1, 12000 )
plt.xlabel(' Fuel Unit ')
```
